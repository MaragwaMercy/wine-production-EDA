I will drop any row that have 2 or 3 values missing of the following three columns because if only one is missing, it can be calculated using the other two.

```python
missing_two_or_more = data[['HarvestedAcres', 'Yield(Unit/Acre)', 'Production']].isnull().sum(axis=1) >= 2

# Step 2: Drop the rows
df = data[~missing_two_or_more]
```

# Data Analysis

```python
yearly_production = df.groupby('Year')['Production'].sum()
yearly_production_sorted = yearly_production.sort_values(ascending=False)
year_highest_production = yearly_production_sorted.idxmax()
highest_production_value = yearly_production_sorted.max()
year_lowest_production = yearly_production_sorted.idxmin()
lowest_production_value = yearly_production_sorted.min()
print(f"Year with highest production: {year_highest_production}, Production: {highest_production_value} Tons")

print(f"Year with lowest production: {year_lowest_production}, Production: {lowest_production_value} Tons")
```

```python
yearly_value = df.groupby('Year')['Value(Dollars)'].sum()
yearly_value_sorted = yearly_value.sort_values(ascending=False)
year_highest_value = yearly_value_sorted.idxmax()
highest_value = yearly_value_sorted.max()
year_lowest_value = yearly_value_sorted.idxmin()
lowest_value = yearly_value_sorted.min()
print(f"Year with highest value is: {year_highest_value}, with Total value of: {highest_value} Dollars")

print(f"Year with lowest value is: {year_lowest_value}, with Total value of: {lowest_value} Dollars")
```

```python
county_production = df.groupby('County')['Production'].sum()
county_production_sorted = county_production.sort_values(ascending=False)
top_five_counties = county_production_sorted.head(5)
print(top_five_counties)
```

```python
county_value = df.groupby('County')['Value(Dollars)'].sum()
county_value_sorted = county_value.sort_values(ascending=False)
top_five_counties = county_value_sorted.head(5)
print(top_five_counties)
```

```python
average_price_per_unit = df.groupby('Year')['Price(Dollars/Unit)'].mean()
top_years_price_per_unit = average_price_per_unit.sort_values(ascending=False).head(5)
print("\nTop 5 Years with Highest Price per Unit:")
print(top_years_price_per_unit)
```

```python
# highest harvested area 
county_highest_harvested_area = df.groupby('Year')['HarvestedAcres'].idxmax()
highest_harvested_area_county = df.loc[county_highest_harvested_area, ['Year', 'County']]
highest_harvested_area_county.columns = ['Year', 'County with Highest Harvested Area']

# highest yield per acre 
county_highest_yield_per_acre = df.groupby('Year')['Yield(Unit/Acre)'].idxmax()
highest_yield_per_acre_county = df.loc[county_highest_yield_per_acre, ['Year', 'County']]
highest_yield_per_acre_county.columns = ['Year', 'County with Highest Yield per Acre']


# highest price per unit 
county_highest_price_per_unit = df.groupby('Year')['Price(Dollars/Unit)'].idxmax()
highest_price_per_unit_county = df.loc[county_highest_price_per_unit, ['Year', 'County']]
highest_price_per_unit_county.columns = ['Year', 'County with Highest Price per Unit']

#  highest production
county_highest_production = df.groupby('Year')['Production'].idxmax()
highest_production_county = df.loc[county_highest_production, ['Year', 'County']]
highest_production_county.columns = ['Year', 'County with Highest Production']

# highest value 
county_highest_value = df.groupby('Year')['Value(Dollars)'].idxmax()
highest_value_county = df.loc[county_highest_value, ['Year', 'County']]
highest_value_county.columns = ['Year', 'County with Highest Value']

# Merge the dataframes on 'Year'
result = highest_harvested_area_county.merge(highest_yield_per_acre_county, on='Year')
result = result.merge(highest_price_per_unit_county, on='Year')
result = result.merge(highest_production_county, on='Year')
result = result.merge(highest_value_county, on='Year')

result
```

# Results


*California Wine Industry Analysis (1980-2020)*

The analysis of California's wine industry spanning over four decades (1980-2020) revealed several significant trends and insights:

**Trends in Wine Production:**

The wine production in California has witnessed notable fluctuations over the years, characterized by distinct spikes and declines.
The year 2006 saw a remarkable decrease in production, despite being the second-highest year in terms of harvested area. This decline led to a surge in the average price per unit, reaching its highest point in the entire dataset.
Wine Production Value:

The year 2007 stood out with the highest production value across the 41-year period, representing a substantial increase from the preceding year (2006).

**County-wise Analysis:**

Napa County emerged as a prominent player in the industry, consistently commanding the highest price per unit for almost a decade (from 2012 to 2020).

Napa County also contributed the highest value in wine production from 2008 to 2020, leading for 30 out of the 41 years in our dataset.

San Joaquin County established itself as a leader in harvested area, maintaining the highest position for 16 consecutive years (from 2005 to 2020) and 24 years overall.

Santa Cruz County experienced a period of dominance in price per ton from 1987 to 1997.

Tulare County excelled in yield per acre from 2013 to 2020.

Fresno County demonstrated remarkable consistency in total production, leading for 32 out of the 41 years.

# Data Limitations

**Data Accuracy and Completeness:** The accuracy and completeness of the data are contingent on the sources used. Any inaccuracies, missing information, or discrepancies in the data could potentially affect the conclusions drawn.

**Limited Scope of Factors Considered:** The project primarily focused on yield, price, and production value. Other factors, such as consumer preferences, climate change impacts, and global market dynamics, could play significant roles but were not included in this analysis.

**Regional Variability:** The analysis focused on a broad county-level view. More granular data at the sub-regional level could provide deeper insights.

**Assumption of Homogeneity:** The analysis assumes that factors impacting one county apply uniformly across the state. In reality, each county may have unique characteristics and challenges not fully captured in this study.

**External Influences not Considered:** External factors like policy changes, economic events, or unforeseen events (e.g., natural disasters) were not factored into the analysis but can have a significant impact on the industry and could explain the trend such as low production in 2006 and the high total production value  in 2007.
